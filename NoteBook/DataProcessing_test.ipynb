```python
import pandas as pd
```

```python
datadanang = "datadanang_unclean.csv"
datahue = "data_Hue_unclean.csv"
```

```python
df = pd.read_csv(datadanang, sep=";")
```

```python
duplicate = df[df.duplicated()] #check duplicate
#duplicate 
df = df.drop_duplicates(subset = 'name') #drop dulicate 
#df
```

```python
df["id"] = df["type"] + "_" + df.index.astype(str).str.zfill(3)

df["full_text"] = ( 
    df["description"].fillna("") + " " +
    df["time"].fillna("") + " " +
    df["price"].fillna("") + " " +
    df["location"].fillna("") + " " +
    df["area"].fillna("") + " " +
    df["note"].fillna("")
)
df["vector_search"] = df["note"].fillna("")
```
